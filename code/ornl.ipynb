```python
# the terminal is being killed, maybe error, use try except and log error

def get_classification_info(laz_path):
    if laz_path:  # Only process valid paths
        try:
            qi, pipeline = pdal_quick_info(laz_path)
            uc = pdal_pc_classes(pipeline)
            nuc = len(uc)  # Number of unique classes
            return uc, nuc
        except Exception as e:
            print(f"Error processing {laz_path}: {e}")
            return None, None
    else:
        return None, None

# Function to parallelize the extraction of uc and nuc for each file
def process_files_parallel(file_paths):
    with ProcessPoolExecutor() as executor:
        return list(executor.map(get_classification_info, file_paths))

# Apply the function to the fullpath column and create 'uc' and 'nuc' columns
fullpaths = df['fullpath'].tolist()
classification_info = process_files_parallel(fullpaths)

# Convert the classification_info into separate 'uc' and 'nuc' columns
df[['uc', 'nuc']] = pd.DataFrame(classification_info, index=df.index)
```
